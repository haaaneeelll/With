```python
import pandas as pd

# CSV 파일을 읽을 때 인코딩을 명시적으로 지정합니다 (CP949 인코딩 사용)
df1 = pd.read_csv("단독다가구(전월세)_실거래가_20230804133002.csv", encoding='cp949')
df1['구분'] = '단독다가구'
df2 = pd.read_csv("단독다가구(전월세)_실거래가_20230804133228.csv", encoding='cp949')
df2['구분'] = '단독다가구'
df3 = pd.read_csv("아파트(전월세)_실거래가_20230804132558.csv", encoding='cp949')
df3['구분'] = '아파트'
df4 = pd.read_csv("아파트(전월세)_실거래가_20230804132652.csv", encoding='cp949')
df4['구분'] = '아파트'
df5 = pd.read_csv("연립다세대(전월세)_실거래가_20230804132834.csv", encoding='cp949')
df5['구분'] = '연립다세대'
df6 = pd.read_csv("연립다세대(전월세)_실거래가_20230804132851.csv", encoding='cp949')
df6['구분'] = '연립다세대'
df7 = pd.read_csv("오피스텔(전월세)_실거래가_20230804133307.csv", encoding='cp949')
df7['구분'] = '오피스텔'
df8 = pd.read_csv("오피스텔(전월세)_실거래가_20230804133331.csv", encoding='cp949')
df8['구분'] = '오피스텔'



# DataFrames를 병합합니다
merged_df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)

# 병합된 DataFrame을 출력합니다
print(merged_df)

# 병합된 DataFrame을 새로운 CSV 파일로 저장합니다 (UTF-8 인코딩 사용)
merged_df.to_csv("2021_2022_강남구전세실거래가.csv", index=False, encoding='utf-8')

```
