```python
import pandas as pd

df_subway = pd.read_csv('역세권.csv',)
df_subway
```

```python
df_subway = df_subway.rename(columns={'위도':'역세권위도', '경도':'역세권경도'})
df_subway.info()
```

```python
df = pd.merge(df_main, df_subway, how='cross')
df
```

```python
from geopy.distance import geodesic
from tqdm import tqdm
import pandas as pd

def calculate_distance(row):
    try:
        return geodesic((row['Latitude'], row['Longitude']), (row['역세권위도'], row['역세권경도'])).kilometers
    except (ValueError, TypeError):
        return None

# Create a new column with calculated distances
tqdm.pandas()  # Enable tqdm on pandas operations
df['역세권 개수'] = df.progress_apply(calculate_distance, axis=1)

df.head()
```

```python
df
```

```python
df['역세권 개수'] = df['역세권 개수'].apply(lambda x: round(x, 2))
```

```python
df.groupby(['index','도로명주소_x'])['역세권 개수'].count()
```

```python
def count_subway(series):
    return (series<=0.5).sum()
    
  

df_groupby = df.groupby(['index','도로명주소_x'])['역세권 개수'].agg(count_subway)
```

```python
df_groupby = pd.DataFrame(df_groupby)
df_groupby
```

```python
df_groupby.reset_index(drop=False, inplace=True)
```

```python
df_groupby
```

```python
(df_main['도로명주소_x']==df_groupby['도로명주소_x']).all()
```

```python
df_main = pd.concat([df_main, df_groupby['역세권 개수']], axis=1)
```
