```python
import pandas as pd

df_main = pd.read_csv('c:/csv/류민석/강남구전세루트데이터_수정.csv', encoding='utf-8')
df_main.head()
```

```python
df_macn = pd.read_csv('c:/csv/맥도날드.csv',encoding='cp949')
df_macn.head()
```

```python
df_macn_test = df_macn.reset_index(drop=False)
```

```python
df_macn_test
```

```python
df_main.reset_index(drop=False, inplace=True)
```

```python
df_main
```

```python
df = pd.merge(df_main, df_macn_test, how='cross')
df
```

```python
import geopy.distance
```

```python
def calculate_distance(row):
     return geopy.distance.distance((row['Latitude'], row['Longitude']), (row['맥도날드 위도'], row['맥도날드 경도'])).km

df['Distance_to_macn'] = df.apply(calculate_distance, axis=1)
```

```python
df
```

```python
df['Distance_to_macn'] = df['Distance_to_macn'].apply(lambda x: round(x, 2))
```

```python
df = df.drop('index_y', axis=1) # 맥날인덱스 열 삭제(민석님꺼에서는 없길래)
```

```python
df = df.rename(columns={'index_x':'index'})
```

```python
df
```

```python
df.to_csv('c:/csv/강남구전세루트데이터+맥날.csv', index=False, encoding='cp949')
```

```python
import pandas as pd

df = pd.read_csv('c:/csv/강남구전세루트데이터+맥날.csv', encoding='cp949')
```

```python
df.groupby(['index','도로명주소'])['Distance_to_macn'].count()
```

```python
def count_macn(series):
    return (series<=0.4).sum()
    
  

df_groupby = df.groupby(['index','도로명주소'])['Distance_to_macn'].agg(count_macn)
```

```python
df_groupby = pd.DataFrame(df_groupby)
df_groupby
```

```python
df_groupby.reset_index(drop=False, inplace=True)
```

```python
df_groupby
```

```python
df_main = pd.read_csv('c:/csv/류민석/강남구전세루트데이터_수정.csv', encoding='utf-8')
df_main.head()
```

```python
(df_main['도로명주소']==df_groupby['도로명주소']).all()
```

```python
df_main = pd.concat([df_main, df_groupby['Distance_to_macn']], axis=1)
```
