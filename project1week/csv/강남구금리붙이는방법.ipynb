```python
df1 = pd.read_csv('강남구전세(3차전처리).csv')
df1.drop(columns='Unnamed: 0', inplace=True)
df1
```

```python
df1.info()
```

```python
df2 = pd.read_csv('금리.csv')
df2
```

```python
df1['계약시작'] = pd.to_datetime(df1['계약시작'])
df1['계약시작년월'] = df1['계약시작'].dt.to_period('M')

df1.head()

```

```python
df2['금리년월일'] = pd.to_datetime(df2['금리년월일'])
df2['금리'] = pd.to_numeric(df2['금리'].str.replace('%', ''))
df2['계약시작년월'] = df2['금리년월일'].dt.to_period('M')


df2.drop(columns ='금리년월일', inplace=True)
df2
```

```python
# 데이터프레임 병합
df1 = df1.merge(df2, on=['계약시작년월'], how='left')
df1
```
