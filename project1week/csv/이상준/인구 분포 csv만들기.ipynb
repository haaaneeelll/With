```python
import pandas as pd

# CSV 파일을 DataFrame으로 읽어오기
pop_data = pd.read_csv('인구밀도.csv', encoding='utf-8')

# 숫자 데이터 열만 선택하여 데이터프레임 구성
numeric_columns = pop_data.columns[3:]
pop_data[numeric_columns] = pop_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# 동별 인구수와 인구밀도를 년도별로 정리
result_data = pd.DataFrame()

for year in range(2015, 2023):
    year_columns = [f'{year}']
    year_data = pop_data[['동별(3)'] + year_columns].copy()
    year_data.rename(columns={f'{year}': '인구수'}, inplace=True)
    result_data[f'{year}년 인구수'] = year_data.groupby('동별(3)').sum()['인구수']

for year in range(2015, 2023):
    year_columns = [f'{year}.{i}' for i in range(1, 3)]
    year_data = pop_data[['동별(3)'] + year_columns].copy()
    year_data.rename(columns={f'{year}.1': '인구밀도'}, inplace=True)
    result_data[f'{year}년 인구밀도'] = year_data.groupby('동별(3)').sum()['인구밀도']

# 결과 출력
result_data
```

```python
result_data.to_csv('인구 분포.csv', index=True)
```
