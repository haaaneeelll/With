```python
df_blue = df_blue.rename(columns={'Latitude':'블루리본위도', 'Longitude':'블루리본경도'})
```

```python
df_blue = df_blue[['블루리본위도','블루리본경도']]
```

```python
def haversine(theta):
    return np.sin((theta)/2)**2

def distance(lat1, lon1, lat2, lon2):
    r = 6371
    lat1 = lat1*np.pi/180
    lon1 = lon1*np.pi/180
    lat2 = lat2.to_numpy()*np.pi/180
    lon2 = lon2.to_numpy()*np.pi/180
    h = haversine(lat2-lat1) + np.cos(lat1)*np.cos(lat2)*haversine(lon2-lon1)
    return 2*r*np.arcsin(h**0.5)
```

```python
import numpy as np

dist_list = []

from tqdm import tqdm
for i in tqdm(range(df.shape[0])):
    df_lat_lon = df.iloc[i][['Latitude','Longitude']]
    dist = distance(df_lat_lon['Latitude'], df_lat_lon['Longitude'], df_blue['블루리본위도'], df_blue['블루리본경도'])
    dist = (dist<=0.5).sum()
    dist_list.append(dist)
```

```python
dist_list = pd.DataFrame(dist_list)
df = pd.concat([df, dist_list], axis=1)
df
```
