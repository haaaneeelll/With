```python
import pandas as pd
import numpy as np

df = pd.read_csv('./통합데이터/강남구전세데이터_찐마지막.csv', encoding='utf-8')
df_subway = pd.read_csv('./subway/강남구지하철역_전_u.csv', encoding='utf-8')
```

```python
df_subway.info()
```

```python
df_subway = df_subway[['위도','경도']]
```

```python
def haversine(theta):
    return np.sin((theta)/2)**2

def distance(lat1, lon1, lat2, lon2):
    r = 6371
    lat1 = lat1*np.pi/180
    lon1 = lon1*np.pi/180
    lat2 = lat2.to_numpy()*np.pi/180
    lon2 = lon2.to_numpy()*np.pi/180
    h = haversine(lat2-lat1) + np.cos(lat1)*np.cos(lat2)*haversine(lon2-lon1)
    return 2*r*np.arcsin(h**0.5)
```

```python
dist_list = []

from tqdm import tqdm
for i in tqdm(range(df.shape[0])):
    df_lat_lon = df.iloc[i][['Latitude','Longitude']]
    dist = distance(df_lat_lon['Latitude'], df_lat_lon['Longitude'], df_subway['위도'], df_subway['경도'])
    dist = (dist<=0.5).sum()
    dist_list.append(dist)
    
dist_list
```
