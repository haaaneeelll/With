```python
import pandas as pd

df_main = pd.read_csv('./main/강남구전세루트데이터_수정.csv', encoding='utf-8')
df_main.head()
```

```python
df_main.info()
```

```python
df_subway = pd.read_csv('./휴게음식점/맥도날드.csv', encoding='cp949')
df_subway.head()
```

```python
df_subway.info()
```

```python
df_subway = df_subway[['맥도날드 위도','맥도날드 경도']]
df_subway.info()
```

```python
df_main.reset_index(drop=False, inplace=True)
```

```python
df_main
```

```python
df = pd.merge(df_main, df_subway, how='cross')
df
```

```python
(df[['index','도로명주소']].duplicated()==False).sum()
##### Good!!!!!!!
```

```python
from geopy.distance import geodesic

def calculate_distance(row):
    try:
        return geodesic((row['Latitude'], row['Longitude']), (row['맥도날드 위도'], row['맥도날드 경도'])).kilometers
    except (ValueError, TypeError):
        return None

df['Distance_to_station'] = df.apply(calculate_distance, axis=1)

df
```

```python
df['Distance_to_station'] = df['Distance_to_station'].apply(lambda x: round(x, 2))
```

```python
df.to_csv('./휴게음식점/강남구전세루트데이터+맥도날드.csv', index=False, encoding='utf-8')
```

```python
df.groupby(['index','도로명주소'])['Distance_to_station'].count()
```

```python
def count_subway(series):
    return (series<=0.5).sum()
    
  

df_groupby = df.groupby(['index','도로명주소'])['Distance_to_station'].agg(count_subway)
```

```python
df_groupby = pd.DataFrame(df_groupby)
df_groupby
```

```python
df_groupby.reset_index(drop=False, inplace=True)
```

```python
df_groupby
```

```python
(df_main['도로명주소']==df_groupby['도로명주소']).all()
```

```python
df_main = pd.concat([df_main, df_groupby['Distance_to_station']], axis=1)
```
