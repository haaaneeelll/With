```python
import pandas as pd

df = pd.read_csv('./통합데이터/메인+금리(8일 최종본).csv', encoding='utf-8')
df.info()
```

```python
df.head()
```

## 동 빼내기

```python
def address_preprocessing(address):
    lists = address.split(' ')
    return lists[-1]

df['행정구'] = df['시군구'].apply(address_preprocessing)
df['행정구']
```

```python
df.head()
```

## 여기까지 솔루션 입니다!!

```python
df_code = pd.DataFrame(df['행정구'].unique(), columns=['행정구'])
df_code.info()
```

```python
df_code
```

```python
df_code = df_code.sort_values('행정구')
df_code
```

```python
df_code.reset_index(drop=True, inplace=True)
df_code
```

```python
df_code.reset_index(drop=False, inplace=True)
df_code
```

```python
df_code.rename(columns={'index':'행정구코드'}, inplace=True)
df_code
```

```python
df_code.to_csv('./통합데이터/행정구코드.csv', index=False, encoding='utf-8')
```
