```python
from selenium import webdriver
import time
import numpy as np

# chromedriver
# 압축해제한 웹드라이버의 경로와 파일명 지정
driver = webdriver.Chrome()

# Load Page
# chrome을 띄워 네이버 블로그 페이지를 연다.
driver.get(url='https://www.bluer.co.kr/search?query=%EA%B0%95%EB%82%A8%EA%B5%AC&isSearchName=false&location=#filter-restaurant')
# driver.get(url='https://www.naver.com/')

time.sleep(5)

from selenium.webdriver.common.by import By
import time

address = []

for i in range(5):
    if i == 0:
        contents = driver.find_elements(By.CSS_SELECTOR,'li.rl-col')
        
        for content in contents:
            
            adds = content.find_elements(By.CSS_SELECTOR,'dd')
            for index, add in enumerate(adds):
                if index%2==1:
                    add_text = add.text
                    address.append(add_text)

        print(i, len(address))
            
        for k in range(2,8):
            a = '[data-lp="{0}"] > a'.format(k)
            driver.find_element(By.CSS_SELECTOR,a).click()

            time.sleep(3)

            contents = driver.find_elements(By.CSS_SELECTOR,'li.rl-col')

            for content in contents:
                adds = content.find_elements(By.CSS_SELECTOR,'dd')
                for index, add in enumerate(adds):
                    if index%2==1:
                        add_text = add.text
                        address.append(add_text)

            print(i , k, len(address))
    
    else:
        for k in range(i*7+1,i*7+8):
            a = '[data-lp="{0}"] > a'.format(k)
            driver.find_element(By.CSS_SELECTOR,a).click()

            time.sleep(3) 

            contents = driver.find_elements(By.CSS_SELECTOR,'li.rl-col')
        
            for content in contents:
                adds = content.find_elements(By.CSS_SELECTOR,'dd')
                for index, add in enumerate(adds):
                    if index%2==1:
                        add_text = add.text
                        address.append(add_text)

            print(i, k, len(address))



print(len(address))
address

driver.close()
```

```python
def address_preprocessing(address):
    if address.find('(') != -1:
        index = address.find('(')
        return address[:index]
    if address.find('(') == -1:
        lists = address.split(' ')
        return ' '.join(lists[0:4])

df_address['도로명주소'] = df_address['도로명주소'].apply(address_preprocessing)
```
