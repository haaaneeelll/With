```python
df_store = pd.read_csv('맥도날드.csv', encoding='utf-8')
df_store = df_store.drop(columns=['맥날 인허가일자','맥날 폐업일자','강남구 맥도날드'])
df_store
```

```python
df = pd.merge(df_main, df_store, how='cross')
```

```python
df.reset_index()
```

```python
from geopy.distance import geodesic
```

```python
from tqdm import tqdm
def calculate_distance(row):
    try:
        return geodesic((row['Latitude'], row['Longitude']), (row['맥도날드 위도'], row['맥도날드 경도'])).kilometers
    except (ValueError, TypeError):
        return None

tqdm.pandas()
df['미쉐린 거리'] = df.progress_apply(calculate_distance, axis=1)
```

```python
df['미쉐린 거리'] = df['미쉐린 거리'].apply(lambda x: round(x, 2))
```

```python
df.to_csv('강남구전세루트데이터+미쉐린거리.csv', index=False, encoding='utf-8')
```

```python
df = df.reset_index(drop=False)
df
```

```python
def count_starbucks(series):
    return (series<=0.4).sum()
    
  

df_groupby = df.groupby(['도로명주소'])['미쉐린 거리'].agg(count_starbucks)
```

```python
df_groupby = pd.DataFrame(df_groupby)
df_groupby
```
