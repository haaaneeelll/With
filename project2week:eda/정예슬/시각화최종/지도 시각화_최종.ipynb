```python
df = pd.read_csv('강남구전세데이터.csv')
```

```python
macn = pd.read_csv('맥도날드.csv', encoding='cp949')
mch = pd.read_csv('미쉐린.csv', encoding='cp949')
st = pd.read_csv('스타벅스.csv', encoding='cp949')
ds = pd.read_csv('백화점.csv')
market = pd.read_csv('시장.csv',encoding='cp949')
school = pd.read_csv('학교.csv',encoding='cp949')
```

```python
import folium
from folium import plugins
from folium.plugins import HeatMap
import json
import geopandas as gpd

m2 = folium.Map(location=[df['Latitude'].mean() , df['Longitude'].mean()], zoom_start=12.5)

heat_df = df[(df['y'] >= df['y'].quantile(0.45)) & (df['y'] >= df['y'].quantile(0.55))][['Latitude', 'Longitude']]

# heat_df = heat_df.dropna(axis=0, subset=['Latitude','Longitude'])

# List comprehension to make out list of lists
heat_data = [[row['Latitude'],row['Longitude']] for index, row in heat_df.iterrows()]

# Plot it on the map
HeatMap(heat_data,
       radius=15).add_to(m2)

# folium.Marker([37.498523, 127.052101], popup='스타벅스1').add_to(m2)
# folium.Marker([37.494525, 127.062584], popup='스타벅스2').add_to(m2)

# 다른 데이터프레임에서 마커 데이터 가져오기
macn_data1 = macn[['맥도날드 위도', '맥도날드 경도', '강남구 맥도날드']].values.tolist()
st_data2 = st[['스타벅스 위도', '스타벅스 경도', '강남구 스타벅스']].values.tolist()
mch_data3 = mch[['미쉐린 위도', '미쉐린 경도', '강남구 미쉐린']].values.tolist()
ds_data4= ds[['Latitude', 'Longitude', '사업장명']].values.tolist()
market_data5= market[['Latitude', 'Longitude', '사업장명']].values.tolist()
school_data6= school[['Latitude', 'Longitude', '학교명']].values.tolist()


for data in macn_data1:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='red',icon='leaf')
    ).add_to(m2)

for data in st_data2:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='green', icon='star')
    ).add_to(m2)
    
for data in mch_data3:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='blue', icon='cloud')
    ).add_to(m2)

for data in ds_data4:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='purple', icon='info-sign')
    ).add_to(m2)
    
for data in market_data5:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='orange', icon='flag')
    ).add_to(m2)

for data in school_data6:
    folium.Marker(
        location=[data[0], data[1]],
        popup=data[2],
        icon=folium.Icon(color='pink', icon='heart')
    ).add_to(m2)
    
# Shapefile 로드
gdf = gpd.read_file('Z_SOP_BND_ADM_DONG_PG.shp')

# GeoJson 레이어 추가
folium.GeoJson(
    gdf,
    style_function=lambda feature: {
        'color': 'black',  # 구분선 색상 설정
        'weight': 2,       # 구분선 굵기 설정
        'fillColor': 'transparent'  # 구분선 내부 색상 설정
    }
).add_to(m2)

# Display the map
m2

#output_html = 'my_map.html'
#m2.save(output_html)

#print(f"지도가 '{output_html}' 파일로 저장되었습니다.")
```
